## 1. Cercles with recursive functions

```python
def circles(n):
    """The following function takes as parameter an integer n>=1 and use recusivity to output n sequence of circles with the
    same center"""
    if n  < 1:
        return 'Be sure that you enter an integer >=1'
    elif n == 1:
        return circle((0,0), n/100, fill=False, color ="black") # The value 100 is a self choice
    else:
        return circle ((0,0), n/100, fill=False,axes=True,color ="black")+circles(n-1)
```

Let see what we get if we call the function

```python
circles(1)
```

```python
circles(3)
```

```python
circles(100)
```

## 2. 3D Circles

A 3d representation allows to have a better appreciation on how those circles are drawn

```python
def circ_3D(n):
    """The following function takes as parameter an integer n>=1 and use recusivity to output n sequence of circles with the
    same center in 3 dimensions space"""
    import random as rd
    if n == 1:
        return circle((0,0,0), (50-n)/50, color= 'black', fill=False,axes=False).translate(0, 0, n/1000)
    else:
        return circle ((0,0,0), (50-n)/50, color= 'black', fill=False,axes=False).translate(0, 0, n/1000)+circ_3D(n-1)
```

## 5.2. The Final interact function

Select the dimension to view and select in a range of type of circles functions the circles to view. $\textbf{But note that Animation might take times to be generated}$ especially 3D animation.

#### N.B: As said above if there is an error when calling a particular circle function, just move back or fourth the value of n

```python
@interact
def intercercles(Dimension=selector(['2D', '3D'],buttons=True),n=(1,50),
             type=['None','circles','circles4','circles2','circles3','circles5','circles6','circles7']   ):
    
    if Dimension == '2D':
        
        if type == 'None':
            return 'Select one of the circles type'
        elif type == 'circles':
            return animate([circles(n) for n in range(1,51)])
        elif type == 'circles4':
            return animate([circles4(n) for n in range(1,51)])
        
        elif type == 'circles2':
            return animate([circles2(n) for n in range(1,51)])
        
        elif type == 'circles3':
            return animate([circles3(n) for n in range(1,51)])
        
        ## We didn't set animation for the following conditions due to depth bugs
        elif type == 'circles5':
            return circles5(n) 
        
        elif type == 'circles6':
            return circles6(n) 
        
        else:
            return circles7(n) 
        
    else:
        if type == 'None':
            return 'Select one of the circles type'
        elif type == 'circles':
            return animate([circ_3D(n) for n in range(1,51)])
        
        elif type == 'circles4':
            return animate([circ_3D4(n) for n in range(1,51)])
        
        elif type == 'circles2':
            return animate([circ_3D2(n) for n in range(1,51)])
        
        elif type == 'circles3':
            return animate([circ_3D3(n) for n in range(1,51)])
        
        ## We didn't set animation for the following conditions due to depth bugs
        elif type == 'circles5':
            return circ_3D5(n)
        
        elif type == 'circles6':
            return circ_3D6(n) 
        
        else:
            return circ_3D_S(n)
        
```
